### Experiment by Chloe for poscon negcon selection
- 2022_08_22_Batch_1

```python
rootDir='/home/ubuntu/bucket/projects/2021_09_01_VarCHAMP/workspace/'
batch='2022_08_22_Batch_1'
platemap=pd.read_csv(rootDir+'/metadata/platemaps/'+batch+'/platemap/2022_08_08_PPL8_100-800.txt', sep='\t')
platemap['batch']=batch
# batch_platemap= pd.read_csv(rootDir+'/metadata/'+batch+'/platemap/BRD4780_1.txt', sep=" ", header=None)
annot_df=platemap.rename(columns={'plate_map_name':'Metadata_Plate','well_position':'Metadata_Well'})

list_of_control_names=['negcon']
annot_df['control']=False
annot_df.loc[annot_df['control_type'].isin(list_of_control_names),'control']=True
annot_df['Variant']=annot_df['allele'].str.upper().str.rstrip('_NA').str.rstrip(' WT').str.replace(' ','_')
# annot_df['Variant']=annot_df['Sample name']
annot_df['Gene']=annot_df['Variant'].apply(lambda x: x.split('_')[0])
annot_df['MT']=annot_df['Variant'].apply(lambda x: x.split('_')[-1])
# annot_df['Metadata_Sample_uniqe']=annot_df['Variant']+'-'+annot_df['Treatment']
annot_df.loc[annot_df['Gene']==annot_df['MT'],'MT']=''

annot_df['Metadata_batch_Plate']=annot_df['batch']+'-'+annot_df['Metadata_Plate']
annot_df['Variant']=annot_df['Variant'].str.replace('_',' ')
annot_df['Metadata_Sample_Unique']=annot_df['Variant']+'-'+annot_df['vTitre'].astype(str)
annot_df.to_csv(save_reprocessed_path+batch+'.csv',index=False)
annot_df.head()
```

### First Pass - Pilot Variant Painting
- 2023_02_23_Batch_3

```python
rootDir='/home/ubuntu/bucket/projects/2021_09_01_VarCHAMP/workspace/'
batch='2023_02_23_Batch_3'
platemap=pd.read_csv(rootDir+'/metadata/platemaps/'+batch+'/platemap/PPL9.txt', sep='\t')
platemap=platemap[~platemap['node_type'].isnull()].reset_index(drop=True)
platemap['batch']=batch
# batch_platemap= pd.read_csv(rootDir+'/metadata/'+batch+'/platemap/BRD4780_1.txt', sep=" ", header=None)
annot_df=platemap.rename(columns={'plate_map_name':'Metadata_Plate','well_position':'Metadata_Well'})
list_of_control_names=['TC']
annot_df['control']=False
annot_df.loc[annot_df['node_type'].isin(list_of_control_names),'control']=True
# annot_df['Variant']=annot_df['allele'].str.upper().str.rstrip('_NA').str.rstrip(' WT').str.replace(' ','_')
# annot_df['Variant']=annot_df['Sample name']

annot_df['Gene']=annot_df['Symbol']#.apply(lambda x: x.split('_')[0])
annot_df['MT']=annot_df['aa_change']#.apply(lambda x: x.split('_')[-1])
# annot_df['Metadata_Sample_uniqe']=annot_df['Variant']+'-'+annot_df['Treatment']
# annot_df.loc[annot_df['Gene']==annot_df['MT'],'MT']=''
annot_df.loc[annot_df['MT'].isnull(),'MT']=''


# annot_df['Variant']=annot_df['Variant'].str.replace('_',' ')
annot_df['Variant']=annot_df['Gene']+' '+annot_df['MT']
annot_df['Variant']=annot_df['Variant'].str.rstrip()
annot_df['Metadata_Sample_Unique']=annot_df['Variant']#+'-'+annot_df['vTitre'].astype(str)

plates = ['2022_10_11_PPL9_P1','2022_09_29_PPL9_P2','2022_09_29_PPL9_P3']
annot_ls=[]
for p in plates:
#     print(p)
    annot_df2 = annot_df.copy()
    annot_df2['Metadata_Plate'] = p
    annot_ls.append(annot_df2)
    
annot_df_all=pd.concat(annot_ls, ignore_index=True)
annot_df_all['Metadata_batch_Plate']=annot_df_all['batch']+'-'+annot_df_all['Metadata_Plate']

annot_df_all.to_csv(save_reprocessed_path+batch+'.csv',index=False)
annot_df_all.head()
```
